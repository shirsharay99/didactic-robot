```python
X = df[['Open','High','Low','Close']]
y = df['Volume']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
sns.scatterplot(y_test,predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot(y_test-predictions, bins =20)
```

```python
sns.lmplot('Open','Volume',data = df, palette='Set1')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logR = LogisticRegression()
```

```python
logR.fit(X_train,y_train)
```

```python
predictions = logR.predict(X_test)
```

```python
sns.scatterplot(y_test,predictions)
```

```python
sns.distplot(y_test-predictions)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
def first_day(entry):
    """
    Returns the first instance of the period, regardless of sampling rate.
    """
    if len(entry):  # handles the case of missing data
        return entry[0]
```

```python
df.resample(rule = 'M').apply(first_day)
```

```python
df.resample(rule = 'A').apply(first_day)
```

```python
df.resample(rule = 'B').apply(first_day)
```
